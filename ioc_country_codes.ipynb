```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_IOC_country_codes"
```

```python
tabellen = pd.read_html(url, storage_options={'User-Agent': 'Mozilla/5.0'})

print(f"Gefundene Tabellen: {len(tabellen)}")
```

```python
df = tabellen[0]
```
